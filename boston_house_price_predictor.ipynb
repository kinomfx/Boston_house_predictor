```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = d['medv']
```

```python
d.drop(columns=['medv','ID'] ,inplace = True)
x=d
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x , y)
```

```python
df = pd.read_csv('boston-housing/test.csv')
index = df['ID']
df.drop(columns=['ID'],inplace = True)
df2 = pd.DataFrame(model.predict(df.to_numpy()),columns=['Medv'])
df2.insert(0,'ID',index)
df2
```
